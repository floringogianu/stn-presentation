# Affine Transformations

```julia
using PyPlot
```

```julia
function fake_matrix(x, y, m)
    if x < 1 || x > size(m,1) || y < 1 || y > size(m, 2)
        return 0
    else
        return m[x,y]
    end
end

function sample(x,y, m)
    x = x + 0.5
    y = y + 0.5
    x1 = floor(x)
    x2 = ceil(x)
    y1 = floor(y)
    y2 = ceil(y)
    val = fake_matrix(x1, y1, m)*(x2-x)*(y2-y) +
          fake_matrix(x2, y1, m)*(x-x1)*(y2-y) +
          fake_matrix(x1, y2, m)*(x2-x)*(y-y1) +
          fake_matrix(x2, y2, m)*(x-x1)*(y-y1)
    return val
end
```

```julia
img = eye(10).+rotr90(eye(10))
img[3,:] = 1
img[:, 3] = 0.75
PyPlot.imshow(img, cmap="gray")
```
